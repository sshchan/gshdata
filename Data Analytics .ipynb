```python
import os

from collections import defaultdict

rootDir = 'Operation Decrypt Isis (GSH)'

file_types = defaultdict(int)
file_sizes = defaultdict(list) 

for dirName, subdirList, fileList in os.walk(rootDir):
    for fname in fileList:
        ext = fname.split('.')[-1].lower()
        full_path = os.path.join(dirName, fname)
        size = os.path.getsize(full_path)

        file_types[ext] += 1 
        file_sizes[ext].append(size/1000.0)
```

```python
file_sizes
```

```python
import matplotlib.pyplot as plt
 
# Data to plot
labels = file_types.keys()
sizes = [file_types[i] for i in labels]
 
# Plot
plt.pie(sizes, labels=labels,
autopct='%1.1f%%', startangle=140)
 
plt.axis('equal')
plt.show()
```

```python
plt.close()

import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt
 
objects = file_types.keys()
x_pos = np.arange(len(objects))
box_height = [file_types[i] for i in objects]
 
plt.bar(x_pos, box_height, align='center', alpha=0.5)
plt.xticks(x_pos, objects, rotation = 80)
plt.ylabel('Count')
plt.title('File Type')
plt.tight_layout()
plt.savefig('pictures/Number of Diff File Types')

plt.show()
```

```python
import numpy as np
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
from ipywidgets import interact

def plot_sizes(key):
    plt.close()
    x = file_sizes[key]
    num_bins = 50
    n, bins, patches = plt.hist(x, num_bins, facecolor='blue', alpha=0.5)

    plt.title('Distribution of .%s File Sizes' %key)
    plt.xlabel('File Size (in KB)')
    plt.ylabel('Count')
    
    plt.savefig('pictures/distribution_%s.png' %key)
    
interact(plot_sizes, key=file_sizes.keys())
```

```python
for k in file_sizes.keys():
    plot_sizes(k)
```
